# MOVIE RECOMMENDATION
GOAL:: Based on given features movies should be recommended to the person based on content and popularity

```python
#importing dependencies

import numpy as np
import pandas as pd
import difflib #to recommend movies to the user which has to be the close match to the name he has given even though with some errors.
from sklearn.feature_extraction.text import TfidfVectorizer #Used to convert object type data to meaningful feature vectors called numerical data
from sklearn.metrics.pairwise import cosine_similarity #it will give similarity score for all the movies with the given movie name and also to recognize all the higest similarity score movies inorder to recommend to the user
```

# Data pre-processing

```python
# loading the data into the pandas dataframe

movies_data = pd.read_csv('movies.csv')
```

```python
# feature selection for recommendation

required_features = ['genres', 'keywords', 'tagline', 'cast', 'director', 'overview']
required_features
```

```python
#Checking for null values

movies_data.isnull().sum()
```

```python
#Replacing null values with empty string

for feature in required_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all required features

combined_features = movies_data['genres'] + movies_data['keywords'] + movies_data['tagline']+ movies_data['cast'] + movies_data['director'] + movies_data['overview'] 
```

```python
combined_features
```

# Feature Extraction

```python
# converting the required features data into the feature vectors 
# with numerical data we can easily find the cosine similarity data

vectorizer = TfidfVectorizer()


```

```python
#Fitting vectors into the object

vectors = vectorizer.fit_transform(combined_features)
```

```python
print(vectors)
```

## Similarity Scores

```python
# obtaining similarity score using cosine similarity

score = cosine_similarity(vectors)
```
